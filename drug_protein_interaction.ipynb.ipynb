```python
from tdc.multi_pred import DTI
data = DTI(name='BindingDB_Kd')
df = data.get_data()
df = df.sample(1000)  # Keep it light for speed
df.head()
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
import numpy as np

def smiles_to_fp(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        return AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=128)
    else:
        return np.zeros(128)

drug_features = df['Drug'].apply(smiles_to_fp)
X_drug = np.array([np.array(list(fp.ToBitString()), dtype=int) for fp in drug_features])
```

```python
import collections

amino_acids = 'ACDEFGHIKLMNPQRSTVWY'

def protein_to_freq(seq):
    freq = collections.Counter(seq)
    return np.array([freq[aa]/len(seq) if aa in freq else 0 for aa in amino_acids])

X_protein = np.array([protein_to_freq(seq) for seq in df['Target']])
```

```python
X = np.concatenate([X_drug, X_protein], axis=1)
y = df['Y'].apply(lambda x: 1 if x < 1000 else 0).values  # Label: 1 = strong binding, 0 = weak
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
def predict_interaction(smiles, sequence, model):
    # Drug fingerprint
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        drug_fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=128)
        X_drug = np.array(list(drug_fp.ToBitString()), dtype=int)
    else:
        return "❌ Invalid SMILES"

    # Protein vector
    freq = collections.Counter(sequence)
    amino_acids = 'ACDEFGHIKLMNPQRSTVWY'
    X_protein = np.array([freq[aa]/len(sequence) if aa in freq else 0 for aa in amino_acids])

    # Combine
    X_input = np.concatenate([X_drug, X_protein]).reshape(1, -1)

    # Predict
    pred = model.predict(X_input)[0]
    return "✅ Interaction likely" if pred == 1 else "❌ No interaction"
```

```python
# Example drug: Aspirin SMILES
drug = "CC(=O)OC1=CC=CC=C1C(=O)O"

# Example protein: Fake short sequence
protein = "MENSDSDEVEKLLKKAGF"

print(predict_interaction(drug, protein, model))
```

```python
drug = "CC/C(=C(/C1=CC=CC=C1)\\C2=CC=C(C=C2)OCCN(C)C)/C3=CC=CC=C3"
protein = "MQEVNQEMESLSSTFYSKLLDP"

print(predict_interaction(drug, protein, model))
```
